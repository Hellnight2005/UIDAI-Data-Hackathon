```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load dataset
data_path = Path(r"C:\Users\Abhijeet\UIDAI-Data-Hackathon\UIDAI-Data-Hackathon\dataset\enrolment.csv")
df = pd.read_csv(data_path)

# Convert date column (DD-MM-YYYY format)
df["date"] = pd.to_datetime(df["date"], dayfirst=True, errors="coerce")

# Filter Maharashtra & Mumbai Suburban
df_mumbai = df[
    (df["state"] == "Maharashtra") &
    (df["district"] == "Mumbai Suburban")
]

# Calculate total enrolment
df_mumbai["total_enrolment"] = (
    df_mumbai["age_0_5"] +
    df_mumbai["age_5_17"] +
    df_mumbai["age_18_greater"]
)

# Filter year (example: 2025)
df_year = df_mumbai[df_mumbai["date"].dt.year == 2025]

# Total enrolment in the year
print("Total Enrolment in 2025:", df_year["total_enrolment"].sum())

# Group by pincode
pincode_total = df_year.groupby("pincode")["total_enrolment"].sum()

# Bar chart
plt.figure()
pincode_total.plot(kind="bar")
plt.xlabel("Pincode")
plt.ylabel("Total Enrolment")
plt.title("UIDAI Enrolment by Pincode â€“ Mumbai Suburban (2025)")
plt.xticks(rotation=45)
plt.show()
```
