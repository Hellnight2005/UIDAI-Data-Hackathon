```python
import pandas as pd
from pathlib import Path

# Full path to the dataset
data_path = Path(r"C:\Users\Abhijeet\UIDAI-Data-Hackathon\UIDAI-Data-Hackathon\dataset\demographics.csv")

# Check if the file exists
if not data_path.exists():
    raise FileNotFoundError(f"{data_path} not found")
    
# Load the dataset
df = pd.read_csv(data_path)
# Preview the first rows

df.head()
```

```python
# 1️⃣ Show all columns
print("Columns in dataset:")
print(df.columns)

# 2️⃣ Show first 5 rows
print("\nFirst 5 rows:")
display(df.head())

# 3️⃣ Check missing values in each column
print("\nMissing values per column:")
print(df.isnull().sum())

# 4️⃣ Optional: Quick summary of data types and non-null counts
print("\nDataset info:")
df.info()

# 5️⃣ Optional: Descriptive statistics for numeric columns
print("\nDescriptive statistics:")
display(df.describe())
```

```python
for col in df.columns:
    num_unique = df[col].nunique()
    num_missing = df[col].isnull().sum()
    print(f"Column: {col} → Unique values: {num_unique}, Missing values: {num_missing}")
```

```python
import matplotlib.pyplot as plt

# Aggregate by month (date already monthly)
monthly_df = df.groupby('date')[['demo_age_5_17', 'demo_age_17_']].sum().reset_index()

# Plot
plt.figure()
plt.plot(monthly_df['date'], monthly_df['demo_age_5_17'], label='Age 5–17')
plt.plot(monthly_df['date'], monthly_df['demo_age_17_'], label='Age 17+')

plt.xlabel('Month')
plt.ylabel('Number of Updates')
plt.title('Monthly Trend of Aadhaar Demographic Updates by Age Group')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# Ensure date is datetime
# -------------------------------
trend_df['date'] = pd.to_datetime(trend_df['date'], dayfirst=True)

# ===============================
# YEARLY TREND
# ===============================
# trend_df['year'] = trend_df['date'].dt.year

# yearly_trend = (
#     trend_df.groupby('year')[['demo_age_5_17', 'demo_age_17_']]
#     .sum()
#     .reset_index()
# )

# plt.figure()
# plt.plot(yearly_trend['year'], yearly_trend['demo_age_5_17'], label='Age 5–17')
# plt.plot(yearly_trend['year'], yearly_trend['demo_age_17_'], label='Age 17+')
# plt.xlabel('Year')
# plt.ylabel('Number of Updates')
# plt.title('Yearly Aadhaar Demographic Update Trend')
# plt.legend()
# plt.tight_layout()
# plt.show()


# ===============================
# QUARTERLY TREND
# ===============================
trend_df['quarter'] = trend_df['date'].dt.to_period('Q')

quarterly_trend = (
    trend_df.groupby('quarter')[['demo_age_5_17', 'demo_age_17_']]
    .sum()
    .reset_index()
)

quarterly_trend['quarter'] = quarterly_trend['quarter'].astype(str)

plt.figure()
plt.plot(quarterly_trend['quarter'], quarterly_trend['demo_age_5_17'], label='Age 5–17')
plt.plot(quarterly_trend['quarter'], quarterly_trend['demo_age_17_'], label='Age 17+')
plt.xlabel('Quarter')
plt.ylabel('Number of Updates')
plt.title('Quarterly Aadhaar Demographic Update Trend')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# ===============================
# MONTHLY TREND (Month Names)
# ===============================
trend_df['month_num'] = trend_df['date'].dt.month
trend_df['month_name'] = trend_df['date'].dt.strftime('%b')

monthly_trend = (
    trend_df.groupby(['month_num', 'month_name'])[['demo_age_5_17', 'demo_age_17_']]
    .sum()
    .reset_index()
    .sort_values('month_num')
)

plt.figure()
plt.plot(monthly_trend['month_name'], monthly_trend['demo_age_5_17'], label='Age 5–17')
plt.plot(monthly_trend['month_name'], monthly_trend['demo_age_17_'], label='Age 17+')
plt.xlabel('Month')
plt.ylabel('Number of Updates')
plt.title('Monthly Aadhaar Demographic Update Trend')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create ratio safely (avoid division by zero)
trend_df['age_ratio'] = trend_df['demo_age_17_'] / (trend_df['demo_age_5_17'] + 1)

# Convert date to month
trend_df['month_num'] = trend_df['date'].dt.month
trend_df['month_name'] = trend_df['date'].dt.strftime('%b')

# Monthly average ratio
ratio_monthly = (
    trend_df.groupby(['month_num', 'month_name'])['age_ratio']
    .mean()
    .reset_index()
    .sort_values('month_num')
)

# Plot
plt.figure()
plt.plot(ratio_monthly['month_name'], ratio_monthly['age_ratio'])
plt.xlabel('Month')
plt.ylabel('Adult-to-Child Update Ratio')
plt.title('Monthly Trend of Adult-to-Child Aadhaar Update Ratio')
plt.tight_layout()
plt.show()
```
