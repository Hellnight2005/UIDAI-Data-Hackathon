```python
import pandas as pd
from pathlib import Path

# Full path to the dataset
data_path = Path(r"C:\Users\Abhijeet\UIDAI-Data-Hackathon\UIDAI-Data-Hackathon\dataset\demographics.csv")

# Check if the file exists
if not data_path.exists():
    raise FileNotFoundError(f"{data_path} not found")
    
# Load the dataset
df = pd.read_csv(data_path)
# Preview the first rows

df.head()
```

```python
# 1️⃣ Show all columns
print("Columns in dataset:")
print(df.columns)

# 2️⃣ Show first 5 rows
print("\nFirst 5 rows:")
display(df.head())

# 3️⃣ Check missing values in each column
print("\nMissing values per column:")
print(df.isnull().sum())

# 4️⃣ Optional: Quick summary of data types and non-null counts
print("\nDataset info:")
df.info()

# 5️⃣ Optional: Descriptive statistics for numeric columns
print("\nDescriptive statistics:")
display(df.describe())
```

```python
for col in df.columns:
    num_unique = df[col].nunique()
    num_missing = df[col].isnull().sum()
    print(f"Column: {col} → Unique values: {num_unique}, Missing values: {num_missing}")
```
